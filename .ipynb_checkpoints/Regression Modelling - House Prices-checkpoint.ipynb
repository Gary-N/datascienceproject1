## Feature selection

Next, feature selection. Eliminating irrelevant variables is vital to increase model accuracy and reduces overfitting. Furthermore, multicollinearity may be an issue. This is when independent variables are correlated, which can reduce the precision of the regression model.

Multicollinearity will probably not affect the accuracy of the model much. However, I may later find it difficult in interpreting the effects of individual features. For example, the linear equation may take the form of **y = Î¸1 + Î¸2 \* x1 + Î¸3 \* x2...**. If x1 and x2 are correlated, then the coefficients (or weightings) Î¸2 and Î¸3 will change close to parallel for changes in either x1 or x2. In other words, it would be difficult to distinguish the individual effects of the features x1 or x2 on the target variable y.

Filling medians for the test set using information from the training set will give overly optimistic test scores (good for Kaggle competitions) but this would not work in production.**


  I can do this by grouping 'LotFrontage' data by 'Neighborhood' and impute the missing values after calculating the median 'LotFrontage' in each neighbourhood:

```python
print('Nulls in training set before filling medians:', dataset.loc[:ntrain]['LotFrontage'].isnull().sum())
print('Nulls in test set before filling medians:', dataset.loc[ntrain:]['LotFrontage'].isnull().sum())

def medianImpute_LotFrontage():
    dataset.loc[:ntrain,'LotFrontage']=dataset.loc[:ntrain].groupby('Neighborhood',sort=False)['LotFrontage'].transform(lambda x: x.fillna(x.median()))
    dataset.loc[ntrain:,'LotFrontage']=dataset.loc[ntrain:].groupby('Neighborhood',sort=False)['LotFrontage'].transform(lambda x: x.fillna(x.median()))

medianImpute_LotFrontage()
print('Nulls in training set after filling medians:', dataset.loc[:ntrain]['LotFrontage'].isnull().sum())
print('Nulls in test set after filling medians:', dataset.loc[ntrain:]['LotFrontage'].isnull().sum())
```

## Linear regression modelling

```python
from sklearn.model_selection import train_test_split
train=dataset[:ntrain]
test=dataset[ntrain:]

y=train['SalePrice'].reset_index(drop=True)
X=train.drop('SalePrice',axis=1,inplace=False)
print(y.shape)
print(X.shape)
```

The training and test sets have now been preprocessed. Now I will apply the machine learning models to the data.

To train the machine learning algorithms, the training set can be split into 4 parts (X_train, X_test, y_train and y_test). Then, using X_train and y_train to to train the regression algorithm, we can predict the values of y_test by applying the results of the algorithm to X_test. In this way, the predicted values of y (*SalePrice*) can be compared with the actual value of y in y_test and measure how well the algorithm performed.

Depending on the size of the dataset, the ratio of training data to test data may vary. For a training set with 1456 entries, a decent train/test split would be 75:25.

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.25,random_state=42)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

One of the simplest models to implement for multiple linear regression is Ordinary Least Squares (OLS). It chooses the parameters of the linear function (e.g. Î¸ for **y = Î¸1 + Î¸2 \* x1 + Î¸3 \* x2... + Î¸n \* xn** as from earlier) that minimises the sum of the squared differences between the linearly appromixated target y and the actual target y.

The method to do this is to 
1. Define the OLS model
2. Fit the model to the training data
3. Predict the test data using the linear equation
4. Use a metric such as $R^2$ or mean squared error (MSE) to measure the error of the model

I will use MAE and RMSE as evaluation metrics. MAE is the mean of the absolute errors and RMSE is the square root of the mean of squared errors. RSME penalises larger errors (which is more practical) and is interpretable as a function of *SalePrice*.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

linreg_model=LinearRegression()
while True: 
    #loop to deal with unknown error on initial run scipt (where it works the 2nd time)
    try: 
        linreg_model.fit(X_train,y_train) 
        break
    except: 
         continue
y_pred=linreg_model.predict(X_test)
print(f'MAE: {mean_absolute_error(y_test,y_pred)}')
print(f'RMSE: {np.sqrt(mean_squared_error(y_test,y_pred))}')
```

For small datasets such as our housing training set, there may be an appreciable random chance in determining the model score. For example, with 25% of the data (25% * 1456 = 364 data points in this case) used as validation data, the model might happen to perform extremely well or by chance extremely inaccurately - depending on these 364 data points.

Therefore, if computational power is sufficient (which it definitely is in this case), it is a good idea to run cross-validation. In cross-validation, the modelling process is applied to subsets of the training dataset. Our dataset can be split into N subsets, and these subsets (called 'folds') will each in turn be used as the validation set and all other sets would be used as training data. In this way, we end up with N measures of model quality which gives a better representation of the model.

Here, I use cross_val_score to 
1. Fit the model. 
2. Train the model using the K-folds splitting strategy. For each iteration, the target value is predicted based on the training data and validated on the validation set.
3. Using a performance metric, the scores for each of the iterations is computed.
4. Taking the mean of these scores gives the final cross validation score.


```python
from sklearn.model_selection import KFold, cross_val_score

def mae_cv(model):
    cv=KFold(5,shuffle=True,random_state=42)
    scores=-1*cross_val_score(model,X,y,cv=cv,scoring='neg_mean_absolute_error')
    return scores

def rmse_cv(model):
    cv=KFold(5,shuffle=True,random_state=42)
    scores=-1*cross_val_score(model,X,y,cv=cv,scoring='neg_root_mean_squared_error')
    return scores

linreg_mae_cv=mae_cv(linreg_model)
linreg_rmse_cv=rmse_cv(linreg_model)
print(f'MAE from each iteration: {linreg_mae_cv}')
print(f'Mean MAE: {linreg_mae_cv.mean()}')
print(f'RMSE from each iteration: {linreg_rmse_cv}')
print(f'Mean RMSE: {linreg_rmse_cv.mean()}')
```

#### Regularisation models 

Now I will implement more effective regression models that use **regularisation**. Regularisation is important because it avoids overfitting (or high variance) the data by penalising the coefficients of features through minimising the *cost function* (which determines the error between predicted and actual values) of the respective machine learning model.

The size of the regularisation parameter in the cost function, commonly called $\lambda$ or $\alpha$, determines how much the features are penalised or, in other words, how much more we want to underfit/generalise the model so that it fits the test data better. A benefit of this is that the models deal with multicollinearity by penalising related features.

In this notebook, I will use Ridge, Lasso, and Elastic Net regression models. These each have their own cost/optimisation functions with a term that includes $\lambda$ as the tuning parameter. 



Here, I examine the optimal regularisation parameter $\alpha$ that minimises the cost function (or in this case MAE). I then plot a graph showing the convex curve of the error for different $\alpha$.

```python
from sklearn.linear_model import Ridge

alphas_Ridge=[0.001, 0.003, 0.007, 0.01, 0.03, 0.07, 0.1, 0.3, 0.7, 1, 3, 7, 10, 30, 70]

ridge_scores_cv=[mae_cv(Ridge(alpha=i)).mean() for i in alphas_Ridge]

def plot_alpha(alphas,model_scores):
    plt.figure(figsize=(10,7))
    plt.plot(alphas,model_scores,ls='-',marker='x',c='r')
    plt.xlabel(r'alpha $\alpha$')
    plt.ylabel('Mean CV score (MAE)')
    plt.show()
    print(f'The alpha in the list corresponding to the lowest CV score is {alphas[model_scores.index(min(model_scores))]}', f'with an MAE of {min(model_scores)}.')

plot_alpha(alphas_Ridge,ridge_scores_cv)
```

Zooming in, the optimal $\alpha$, for the **Ridge** regression model turns out to be around 15-16:

```python
alphas_Ridge=np.linspace(15,16,30)
ridge_scores_cv=[mae_cv(Ridge(alpha=i)).mean() for i in alphas_Ridge]

plot_alpha(alphas_Ridge,ridge_scores_cv)
```

The optimal $\alpha$ for the **Lasso** regression model is a lot lower; around 1e-4 to 1e-3.

```python
from sklearn.linear_model import Lasso 

alphas_Lasso=np.linspace(0.0001,0.001,30)
lasso_scores_cv=[mae_cv(Lasso(alpha=i, max_iter=5000)).mean() for i in alphas_Lasso]

plot_alpha(alphas_Lasso,lasso_scores_cv)
```

Similarly, the optimal $\alpha$ for the **ElasticNet** regression model is around 1e-4 to 1e-3.

```python
from sklearn.linear_model import ElasticNet
alphas_ElasticNet=np.linspace(0.0005,0.001,50)
ElasticNet_scores_cv=[mae_cv(ElasticNet(alpha=i,max_iter=5000)).mean() for i in alphas_ElasticNet]

plot_alpha(alphas_ElasticNet,ElasticNet_scores_cv)
```

## XGBoost model

I will now experiment with an ensemble machine learning algorithm called XGBoost (extreme gradient boosting). This is similar to a Random Forest algorithm in that it produces a set of decision trees. Gradient boosting starts with a single model in the ensemble and new models are added iteratively through a cycle where the initial model:

1. Generates predictions from all previous models in the ensemble.
2. Calculates the loss/cost function .
3. Uses gradient descent on the loss function to fit a new model with these new calculated parameters.
4. Adds this new model to the ensemble.



For the XGBoost model, there are a few hyperparameters that usually have a significant impact on performance. These are the ones I will focus on tuning. It is possible to use techniques such as grid search or Bayesian optimization to obtain an optimal set of parameters. 

Furthermore, on large datasets, it would be a good idea to set njobs=-1, which uses parallelism i.e. parallel computing using all the processors available. This splits the tasks between cores for greater efficiency. However, this is not necessary because of the size of our dataset.

I will start with the parameters *n_estimators*, *early_stopping_rounds*. *n_estimators* corresponds to the number of cycles (or models in the ensemble) described above. Like regularisation, getting this value right is important as to not underfit or overfit the data. By calculating the performance of the overall model after each cycle/tree, *early_stopping_rounds* helps to find the best number of cycles by finding out when the performance has not improved after N cycles. N is the value of *early_stopping_rounds* and causes the algorithm to choose the cycle, just before the 1st stopping round, for *n_estimators*.

Here, I train the model to find this parameter. Note: *n_estimators* parameter may be *num_boost_round* depending on the xgb function.

#### Base model - default parameters

```python
import xgboost as xgb

#I define some chosen parameters to fiddle with later and their default values
params={
    #booster parameters
    'max_depth':6,
    'min_child_weight':1,
    'subsample':1,
    'colsample_bytree':1,
    'eta':0.3,
    #learning task parameters
    'objective':'reg:squarederror',
    'eval_metric':'rmse'
}

#XGBoost package uses DMatrices instead of numpy arrays or pandas dataframes
dtrain=xgb.DMatrix(X_train, label=y_train)
dtest=xgb.DMatrix(X_test, label=y_test)

xgboost_model_train=xgb.train(params, dtrain, num_boost_round=1000, early_stopping_rounds=5, evals=[(dtest,'test')])
```

```python
print('\n', f'RMSE: {xgboost_model_train.best_score}')
print(f'Optimal number of estimators: {xgboost_model_train.best_iteration}')
```

#### Cross-validation

I will run 5-fold cross validation on the training dataset via the XGBoost cv() function. This will allow me to tune the hyperparameters chosen in params and will give a mean score. It is essential to set a seed parameter so that the scores are based off of the same folds.

```python
xgboost_score_cv=xgb.cv(params, dtrain, num_boost_round=1000, early_stopping_rounds=5,
                    nfold=5, seed=42, metrics={'rmse'})
xgboost_score_cv
```

```python
print(f'Number of trees: {xgboost_score_cv["test-rmse-mean"].argmin()}')
print(f'RMSE for this cross-validation model: {xgboost_score_cv["test-rmse-mean"].min()}')
```

It is seen that the RMSE for the cross validation model is lower than the single validation set model.

#### Tree architecture parameters - max_depth & min_child_weight 

I will now tune the parameters *max_depth* and *min_child_weight* whilst minimising the RMSE. *max_depth* controls the maximum depth of any single tree. Deeper trees find deeper connections in the data, but may overfit for larger depths. *min_child_weight* is the minimum weight needed to create a new leaf node in the tree, where the weight is the addition of n number of samples' weights and is correlated with *max_depth* in that a lower *min_child_weight* allows for children to be created with less weights - increasing complexity and potential overfitting.

I will use a grid search approach to find the optimal pairing of these two parameters.

```python
gridsearch_params=[(max_depth, min_child_weight) for max_depth in range(3,9) for min_child_weight in range(3,9)]

rmse_optimal=float('inf')
params_optimal=None
no_cycles=0
for max_depth, min_child_weight in gridsearch_params:
    params['max_depth']=max_depth
    params['min_child_weight']=min_child_weight
    
    xgboost_score_cv=xgb.cv(params, dtrain, num_boost_round=1000, early_stopping_rounds=5,
                            nfold=5, seed=42,metrics={'rmse'})
    
    xgboost_mean_score_cv=xgboost_score_cv['test-rmse-mean'].min()
    cycles=xgboost_score_cv['test-rmse-mean'].argmin()
    
    print('max_depth={}, min_child_weight={}: RMSE={} for {} cycles'.format(max_depth, min_child_weight, xgboost_mean_score_cv, cycles))
    
    if xgboost_mean_score_cv < rmse_optimal:
        rmse_optimal=xgboost_mean_score_cv
        params_optimal=max_depth, min_child_weight
        no_cycles=cycles

print('Optimal parameters are: max_depth={}, min_child_weight={} with RMSE: {} for {} cycles'.format(params_optimal[0], params_optimal[1], rmse_optimal, no_cycles))
```

```python
params['max_depth']=params_optimal[0]
params['min_child_weight']=params_optimal[1]
```

#### Dataset sampling parameters - subsample & colsample_bytree

subsample controls what percentage of samples are used at each cycle or node. Similarly, colsample_bytree controls what percentage of features are used at each step. The defaults are set to 1 which uses all rows and columns in the training set.

Again I will perform a grid search to find the optimal parameters.

```python
gridsearch_params=[(subsample, colsample_bytree) for subsample in np.linspace(0.6,1,5) for colsample_bytree in np.linspace(0.6,1,5)]

rmse_optimal=float('inf')
params_optimal=None
no_cycles=0
for subsample, colsample_bytree in gridsearch_params:
    params['subsample']=subsample
    params['colsample_bytree']=colsample_bytree
    
    xgboost_score_cv=xgb.cv(params, dtrain, num_boost_round=1000, early_stopping_rounds=5,
                            nfold=5, seed=42,metrics={'rmse'})
    
    xgboost_mean_score_cv=xgboost_score_cv['test-rmse-mean'].min()
    cycles=xgboost_score_cv['test-rmse-mean'].argmin()
    
    print('subsample={}, colsample_bytree={}: RMSE={} for {} cycles'.format(subsample, colsample_bytree, xgboost_mean_score_cv, cycles))
    
    if xgboost_mean_score_cv < rmse_optimal:
        rmse_optimal=xgboost_mean_score_cv
        params_optimal=subsample, colsample_bytree
        no_cycles=cycles

print('Optimal parameters are: subsample={}, colsample_bytree={} with RMSE: {} for {} cycles'.format(params_optimal[0], params_optimal[1], rmse_optimal, no_cycles))
```

```python
params['subsample']=params_optimal[0]
params['colsample_bytree']=params_optimal[1]
```

The default parameters turn out to be the most optimal in minimising the cost function in this case.

#### Learning rate parameter

Finally, *learning_rate* multiplies the predictions of each model by a small number - the learning rate - to increase accuracy in the model. A low value (along with high *n_estimators*) would give a more accurate model but would take longer to train since there would be more iterations.

Note: the *learning_rate* parameter is defined as *eta* depending on the xgb function.

```python
rmse_optimal=float('inf')
params_optimal=None

learning_rates=[0.4, 0.3, 0.2, 0.1, 0.05, 0.01, 0.005]
for i in learning_rates:
    params['eta']=i

    %time xgboost_score_cv=xgb.cv(params, dtrain, num_boost_round=1000, early_stopping_rounds=5, nfold=5, seed=42,metrics={'rmse'})

    xgboost_mean_score_cv=xgboost_score_cv['test-rmse-mean'].min()
    cycles=xgboost_score_cv['test-rmse-mean'].argmin() 

    print('Learning rate={}: RMSE={} for {} cycles'.format(i, xgboost_mean_score_cv, cycles))

    if xgboost_mean_score_cv < rmse_optimal:
        rmse_optimal=xgboost_mean_score_cv
        params_optimal=i
        
print('Optimal learning rate parameter = {} with RMSE: {}'.format(params_optimal, rmse_optimal))
```

It seems that the model converges well between 0.3 and 0.01. When the learning rate is 0.005, it seems that the model gets stuck in a local minimum. 

I will use a learning rate of 0.05 as the RMSE is the most optimal and does not take a large time to compute (~20 secs).

```python
params['eta']=params_optimal
params
```

Training the model with the updated parameters for cross-validation:

```python
xgboost_model_train=xgb.train(params, dtrain, num_boost_round=1000, early_stopping_rounds=5, evals=[(dtest,'test')])
```

```python
print(f'RMSE: {xgboost_model_train.best_score}')
print(f'Optimal number of estimators: {xgboost_model_train.best_iteration}')
```

#### Define final XGBoost model

At each step of parameter tuning, there has been a significant improvement in the performance of the model (apart from the dataset sampling parameters). Depending on the performance, the XGBoost model may be the most useful model for predicting housing prices.

The best model can now be fit to the data to predict *SalePrice*. I will stick with the low level API xgb.train instead of the wrapper xgb.XGBRegressor since it has a useful function that outputs the feature importance by measuring their F-scores.

```python
num_boost_round=xgboost_model_train.best_iteration
xgboost_model=xgb.train(params, dtrain, num_boost_round=num_boost_round, evals=[(dtest, "Test")], verbose_eval=False)

plt.rcParams['figure.figsize']=(8,9)
xgb.plot_importance(xgboost_model, max_num_features=20)
plt.show()
```

## Putting it all together using pipelines

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import LassoCV
from sklearn.linear_model import ElasticNetCV
```

For reference, these are the parameters found from earlier:

```python
alphas_Ridge=np.linspace(15,16,50)
alphas_Lasso=np.linspace(0.0001,0.001,50)
alphas_ElasticNet=np.linspace(0.0005,0.001,50)

xgboost_model=xgb.XGBRegressor(n_estimators=num_boost_round, max_depth=params['max_depth'], min_child_weight=params['min_child_weight'],
                               learning_rate=params['eta'], objective='reg:squarederror')
```

I will use the preprocessing function *RobustScaler* to normalise and standardise the dataset so to improve the accuracy of the machine learning model. This is because the Ridge, Lasso, and Elastic Net regression models are not invariant (unlike OLS). Using a pipeline, this preprocessing step and the modelling step can be performed easier. 

```python
ridge=Pipeline([('scl', RobustScaler()), ('reg', RidgeCV(alphas=alphas_Ridge))])
lasso=Pipeline([('scl', RobustScaler()), ('reg', LassoCV(alphas=alphas_Lasso,max_iter=5000))])
elastic_net=Pipeline([('scl', RobustScaler()), ('reg', ElasticNetCV(alphas=alphas_ElasticNet))])
xgboost=Pipeline([('scl', RobustScaler()), ('reg', xgboost_model)])
```

Calculating the base models performances:

```python
print(f'Mean RMSE for Ridge: {rmse_cv(ridge).mean()}')
print(f'Mean RMSE for Lasso: {rmse_cv(lasso).mean()}')
print(f'Mean RMSE for Elastic Net: {rmse_cv(elastic_net).mean()}')
print(f'Mean RMSE for XGBoost: {rmse_cv(xgboost_model).mean()}')
```

It is seen that Lasso and Elastic Net models perform the best in prediction for cross-validation.
